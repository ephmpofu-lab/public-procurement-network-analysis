```python
project_df = austria_df[
    [
        "ID_NOTICE_CAN",
        "ID_LOT",
        "ID_AWARD",
        "YEAR",
        "DT_DISPATCH",
        "CAE_NAME",
        "CAE_TOWN",
        "WIN_NAME",
        "WIN_TOWN",
        "WIN_COUNTRY_CODE",
        "TYPE_OF_CONTRACT",
        "CPV",
        "VALUE_EURO"
    ]
].copy()

print("Rows:", project_df.shape[0])
print("Columns:", project_df.shape[1])
print("Exact duplicate rows:", project_df.duplicated().sum())
```

```python
project_df["CAE_NAME_CLEAN"] = (
    project_df["CAE_NAME"]
    .str.strip()
    .str.lower()
    .str.replace(r"\s+", " ", regex=True)
)

project_df["WIN_NAME_CLEAN"] = (
    project_df["WIN_NAME"]
    .str.strip()
    .str.lower()
    .str.replace(r"\s+", " ", regex=True)
)
```

```python
project_df[
    [
        "CAE_NAME",
        "CAE_NAME_CLEAN",
        "WIN_NAME",
        "WIN_NAME_CLEAN"
    ]
].head(20)
```

```python
print("Unique contracting authorities:",
      project_df["CAE_NAME_CLEAN"].nunique())

print("Unique winning companies:",
      project_df["WIN_NAME_CLEAN"].nunique())
```

```python
project_df.groupby(
    ["CAE_NAME_CLEAN", "WIN_NAME_CLEAN"]
).size().sort_values(ascending=False).head(20)
```

```python
suspicious = project_df[
    project_df["WIN_NAME_CLEAN"].str.contains(
        "partner|rahmenvereinbarung",
        case=False,
        na=False
    )
]

suspicious["WIN_NAME_CLEAN"].value_counts().head(30)
```

```python
placeholder_patterns = [
    r"^rv-partner",
    r"^rahmenvereinbarungspartner$",
    r"keine bekanntgabe",
    r"nicht offen gelegt",
    r"identität.*nicht",
    r"identitäten.*nicht",
    r"rahmenvereinbarung mit mehreren partnern",
    r"es wurden mit .* firmen"
]

pattern = "|".join(placeholder_patterns)

project_df["IS_PLACEHOLDER_WINNER"] = (
    project_df["WIN_NAME_CLEAN"]
    .str.contains(pattern, regex=True, na=False)
)

print(
    project_df["IS_PLACEHOLDER_WINNER"].value_counts()
)
```
