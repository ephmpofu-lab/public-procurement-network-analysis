```python
project_df = austria_df[
    [
        "ID_NOTICE_CAN",
        "ID_LOT",
        "ID_AWARD",
        "YEAR",
        "DT_DISPATCH",
        "CAE_NAME",
        "CAE_TOWN",
        "WIN_NAME",
        "WIN_TOWN",
        "WIN_COUNTRY_CODE",
        "TYPE_OF_CONTRACT",
        "CPV",
        "VALUE_EURO"
    ]
].copy()

print("Rows:", project_df.shape[0])
print("Columns:", project_df.shape[1])
print("Exact duplicate rows:", project_df.duplicated().sum())
```

```python
project_df["CAE_NAME_CLEAN"] = (
    project_df["CAE_NAME"]
    .str.strip()
    .str.lower()
    .str.replace(r"\s+", " ", regex=True)
)

project_df["WIN_NAME_CLEAN"] = (
    project_df["WIN_NAME"]
    .str.strip()
    .str.lower()
    .str.replace(r"\s+", " ", regex=True)
)
```

```python
placeholder_patterns = [
    r"^rv-partner",
    r"^rahmenvereinbarungspartner$",
    r"keine bekanntgabe",
    r"nicht offen gelegt",
    r"identität.*nicht",
    r"identitäten.*nicht",
    r"rahmenvereinbarung mit mehreren partnern",
    r"es wurden mit .* firmen"
]

pattern = "|".join(placeholder_patterns)

project_df["IS_PLACEHOLDER_WINNER"] = (
    project_df["WIN_NAME_CLEAN"]
    .str.contains(pattern, regex=True, na=False)
)

print(
    project_df["IS_PLACEHOLDER_WINNER"].value_counts()
)
```

```python
network_df = project_df[
    (project_df["IS_PLACEHOLDER_WINNER"] == False) &
    (project_df["CAE_NAME_CLEAN"].notna()) &
    (project_df["WIN_NAME_CLEAN"].notna())
].copy()

print("Original records:", len(project_df))
print("Network-ready records:", len(network_df))
print("Authorities:", network_df["CAE_NAME_CLEAN"].nunique())
print("Companies:", network_df["WIN_NAME_CLEAN"].nunique())
```

```python
edges_df = (
    network_df
    .groupby(
        ["CAE_NAME_CLEAN", "WIN_NAME_CLEAN"],
        as_index=False
    )
    .agg(
        number_of_awards=("ID_AWARD", "nunique"),
        total_value_euro=("VALUE_EURO", "sum")
    )
)

print("Unique authority-company relationships:", len(edges_df))

edges_df.head(10)
```

```python
import networkx as nx

G = nx.from_pandas_edgelist(
    edges_df,
    source="CAE_NAME_CLEAN",
    target="WIN_NAME_CLEAN",
    edge_attr=["number_of_awards", "total_value_euro"],
    create_using=nx.DiGraph()
)

print("Number of nodes:", G.number_of_nodes())
print("Number of edges:", G.number_of_edges())
```

```python
# Check whether any organisation appears as both a contracting authority and a winning company

authorities = set(network_df["CAE_NAME_CLEAN"].unique())
companies = set(network_df["WIN_NAME_CLEAN"].unique())

overlap = authorities.intersection(companies)

print("Authorities:", len(authorities))
print("Companies:", len(companies))
print("Organisations appearing in both roles:", len(overlap))

list(overlap)[:20]
```

```python
# Create unique node IDs so authorities and companies remain separate network node types

edges_df["authority_node"] = "AUTH::" + edges_df["CAE_NAME_CLEAN"]
edges_df["company_node"] = "COMP::" + edges_df["WIN_NAME_CLEAN"]

edges_df[
    [
        "authority_node",
        "company_node",
        "number_of_awards",
        "total_value_euro"
    ]
].head(10)
```

```python
# Rebuild the directed network using separate authority and company node IDs

G = nx.from_pandas_edgelist(
    edges_df,
    source="authority_node",
    target="company_node",
    edge_attr=["number_of_awards", "total_value_euro"],
    create_using=nx.DiGraph()
)

print("Number of nodes:", G.number_of_nodes())
print("Number of edges:", G.number_of_edges())
```

```python
# Add node type attributes to distinguish contracting authorities from winning companies

for node in G.nodes():

    if node.startswith("AUTH::"):
        G.nodes[node]["node_type"] = "authority"

    elif node.startswith("COMP::"):
        G.nodes[node]["node_type"] = "company"


# Check that node types were assigned correctly

from collections import Counter

node_type_counts = Counter(
    nx.get_node_attributes(G, "node_type").values()
)

print(node_type_counts)
```

```python
# Add node type attributes to distinguish contracting authorities from winning companies

for node in G.nodes():

    if node.startswith("AUTH::"):
        G.nodes[node]["node_type"] = "authority"

    elif node.startswith("COMP::"):
        G.nodes[node]["node_type"] = "company"


# Check that node types were assigned correctly

from collections import Counter

node_type_counts = Counter(
    nx.get_node_attributes(G, "node_type").values()
)

print(node_type_counts)
```

```python
# Identify companies connected to the largest number of contracting authorities

company_nodes = [
    node for node in G.nodes()
    if node.startswith("COMP::")
]

company_authority_count = {
    company: G.in_degree(company)
    for company in company_nodes
}

top_companies = sorted(
    company_authority_count.items(),
    key=lambda x: x[1],
    reverse=True
)[:20]

top_companies
```

```python
# Identify contracting authorities connected to the largest number of winning companies

authority_nodes = [
    node for node in G.nodes()
    if node.startswith("AUTH::")
]

authority_supplier_count = {
    authority: G.out_degree(authority)
    for authority in authority_nodes
}

top_authorities = sorted(
    authority_supplier_count.items(),
    key=lambda x: x[1],
    reverse=True
)[:20]

top_authorities
```

```python
# Calculate the total procurement value connected to each winning company

company_total_value = {
    company: G.in_degree(
        company,
        weight="total_value_euro"
    )
    for company in company_nodes
}

top_companies_by_value = sorted(
    company_total_value.items(),
    key=lambda x: x[1],
    reverse=True
)[:20]

top_companies_by_value
```

```python
# Identify authority-company relationships with the highest number of awards

top_repeat_relationships = edges_df.sort_values(
    by="number_of_awards",
    ascending=False
)[
    [
        "CAE_NAME_CLEAN",
        "WIN_NAME_CLEAN",
        "number_of_awards",
        "total_value_euro"
    ]
].head(20)

top_repeat_relationships
```

```python
# Calculate degree centrality for all organisations in the procurement network

degree_centrality = nx.degree_centrality(G)

top_degree_centrality = sorted(
    degree_centrality.items(),
    key=lambda x: x[1],
    reverse=True
)[:20]

top_degree_centrality
```

```python
# Calculate betweenness centrality to identify organisations that act as structural bridges

G_undirected = G.to_undirected()

betweenness = nx.betweenness_centrality(G_undirected)

top_betweenness = sorted(
    betweenness.items(),
    key=lambda x: x[1],
    reverse=True
)[:20]

top_betweenness
```

```python
# Calculate PageRank to identify prominent companies based on the importance of their connections

pagerank = nx.pagerank(
    G,
    weight="number_of_awards"
)

top_pagerank = sorted(
    pagerank.items(),
    key=lambda x: x[1],
    reverse=True
)[:20]

top_pagerank
```

```python
# Detect communities before assigning community IDs

G_undirected = G.to_undirected()

communities = nx.community.louvain_communities(
    G_undirected,
    weight="number_of_awards",
    seed=42
)

print("Number of communities detected:", len(communities))
```

```python
# Assign a community ID to every organisation in the network

community_map = {}

for community_id, community in enumerate(communities):
    for node in community:
        community_map[node] = community_id

nx.set_node_attributes(
    G,
    community_map,
    "community"
)

print("Nodes assigned to communities:", len(community_map))
print("Total network nodes:", G.number_of_nodes())
```

```python
# Select the largest detected community for a readable network visualisation

largest_community = max(communities, key=len)

G_largest = G.subgraph(largest_community).copy()

print("Nodes in largest community:", G_largest.number_of_nodes())
print("Edges in largest community:", G_largest.number_of_edges())
```

```python
# Visualise the largest procurement network community

import matplotlib.pyplot as plt

plt.figure(figsize=(16, 12))

pos = nx.spring_layout(
    G_largest,
    seed=42,
    k=0.5
)

nx.draw_networkx_nodes(
    G_largest,
    pos,
    node_size=80,
    alpha=0.8
)

nx.draw_networkx_edges(
    G_largest,
    pos,
    alpha=0.25,
    arrows=False
)

plt.title("Largest Community in the Austrian Public Procurement Network")
plt.axis("off")
plt.tight_layout()
plt.show()
```

```python
# Create an interactive network visualisation of the largest procurement community

from pyvis.network import Network

net = Network(
    height="800px",
    width="100%",
    bgcolor="#ffffff",
    font_color="#000000",
    directed=True
)

net.from_nx(G_largest)

print("Nodes added:", len(net.nodes))
print("Edges added:", len(net.edges))
```

```python
# Build an informative interactive procurement network

from pyvis.network import Network

net = Network(
    height="900px",
    width="100%",
    bgcolor="#ffffff",
    font_color="#222222",
    directed=True,
    select_menu=True,
    filter_menu=True
)

for node in G_largest.nodes():

    role = "Contracting Authority" if node.startswith("AUTH::") else "Winning Company"

    degree = G.degree(node)

    display_name = node.replace("AUTH::", "").replace("COMP::", "")

    net.add_node(
        node,
        label=display_name,
        title=(
            f"<b>{display_name}</b><br>"
            f"Role: {role}<br>"
            f"Connections: {degree}<br>"
            f"Community: {G.nodes[node].get('community', 'N/A')}<br>"
            f"PageRank: {pagerank.get(node, 0):.6f}"
        ),
        group=role,
        size=10 + degree
    )

for source, target, data in G_largest.edges(data=True):

    net.add_edge(
        source,
        target,
        title=(
            f"Awards: {data.get('number_of_awards', 0)}<br>"
            f"Total value: €{data.get('total_value_euro', 0):,.2f}"
        ),
        value=data.get("number_of_awards", 1)
    )

net.show_buttons(filter_=["physics"])



print("Enhanced interactive graph created.")

# Add fixed legend nodes to explain the network colours

net.add_node(
    "LEGEND_AUTHORITY",
    label="Contracting Authority",
    group="Contracting Authority",
    shape="dot",
    size=20,
    physics=False,
    x=-900,
    y=-700,
    fixed=True
)

net.add_node(
    "LEGEND_COMPANY",
    label="Winning Company",
    group="Winning Company",
    shape="dot",
    size=20,
    physics=False,
    x=-900,
    y=-640,
    fixed=True
)

net.write_html("procurement_network_interactive.html")
```
